```python
import random

# List of words to choose from
word_list = ["Lion", "Tiger", "Bear", "Elephant", "Giraffe", "Zebra", "Rhino", "Hippopotamus", "Gorilla", "Chimpanzee", "Monkey", "Koala", "Kangaroo", "Platypus", "Crocodile", "Alligator", "Turtle", "Snake", "Lizard", "Frog", "Toad", "Salamander", "Fish", "Shark", "Whale", "Dolphin", "Octopus", "Squid", "Crab", "Lobster", "Shrimp", "Clam", "Oyster", "Jellyfish", "Starfish", "Snail", "Slug", "Butterfly", "Moth", "Bee", "Ant", "Grasshopper", "Cricket", "Spider", "Scorpion", "Bat", "Owl", "Eagle", "Hawk", "Penguin"]

# Select a random word from the list
chosen_word = random.choice(word_list)

# Create a list to store the guessed letters
guessed_letters = []

# Create a variable to keep track of the number of incorrect guesses
incorrect_guesses = 0

# Create a list to store the letters of the chosen word
word_letters = list(chosen_word)

# Create a list to store the correctly guessed letters
display = ["_" for _ in word_letters]

# Function to print the hangman
def print_hangman():
    print("  ____")
    print(" |    |")
    print(" |    " + ("O" if incorrect_guesses >= 1 else ""))
    print(" |   " + ("/|\\" if incorrect_guesses >= 2 else ""))
    print(" |    " + ("|" if incorrect_guesses >= 3 else ""))
    print(" |   " + ("/ \\" if incorrect_guesses >= 4else ""))
    print("_|_")

# Function to print the current state of the game
def print_game_state():
    print("\n" + " ".join(display))
    print("\nGuessed letters: " + ", ".join(guessed_letters))
    print("Incorrect guesses: " + str(incorrect_guesses))
    print_hangman()

# Main game loop
while True:
    print_game_state()

    # Check if the player has won
    if "_" not in display:
        print("\nCongratulations, you won!")
        break

    # Get the player's guess
    guess = input("\nGuess a letter: ").lower()

    # Check if the guess has already been made
    if guess in guessed_letters:
        print("You have already guessed that letter. Please try again.")
        continue

    # Add the guess to the guessed letters list
    guessed_letters.append(guess)

    # Check if the guess is in the word
    if guess in word_letters:
        # Update the display with the correctly guessed letter
        for i in range(len(word_letters)):
            if word_letters[i] == guess:
                display[i] = guess
    else:
        # Increment the number of incorrect guesses
        incorrect_guesses += 1
        print("Incorrect guess. Please try again.")

    # Check if the player has lost
    if incorrect_guesses == 5:
        print("\nSorry, you lost. The word was " + chosen_word + ".")
        break
```
